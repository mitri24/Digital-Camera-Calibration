```python
#Bibliotheken importieren
import cv2
import numpy as np
import glob
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
# Liste aller Dunkelbilder im Ordner "Bildaufnahmen"
dunkel_files = sorted(glob.glob("Bildaufnahmen/Dunkel*.png"))

# Überprüfe, ob Dunkelbilder gefunden wurden
if not dunkel_files:
    print("Keine Dunkelbilder gefunden.")
else:
    print(f"{len(dunkel_files)} Dunkelbilder gefunden.")

# Erstelle eine Liste der Grauwertbilder
bilder = []
for file in dunkel_files:
    img = cv2.imread(file)
    if img is None:
        print(f"Fehler beim Laden des Bildes: {file}")
        continue
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY).astype("float32")
    bilder.append(gray)

# Überprüfe, ob Bilder geladen wurden
if len(bilder) == 0:
    print("Keine gültigen Bilder zum Berechnen des Mittelwerts.")
else:
    # Mittelwert der Dunkelbilder berechnen
    dunkelbild = np.mean(bilder, axis=0)

    # Überprüfe die Form des Mittelwerts
    print(f"Form des Mittelwerts: {dunkelbild.shape}")

    # Dunkelbild speichern im Ordner "BildOuput"
    np.save("BildOutput/Dunkelbild.npy", dunkelbild)

    # Dunkelbild kontrastmaximiert darstellen
    plt.imshow(dunkelbild, cmap="gray")
    plt.title("Kontrastmaximiertes Dunkelbild")
    plt.show()
```

```python
# Weißbilder einlesen und Mittelwert berechnen

# Liste aller Weißbilder im Ordner "Bildaufnahmen"
weiss_files = sorted(glob.glob("Bildaufnahmen/Hell*.png"))

# Überprüfen, ob Bilder im Ordner vorhanden sind
if not weiss_files:
    print("Keine Weißbilder gefunden.")
else:
    print(f"{len(weiss_files)} Weißbilder gefunden.")

# Erstelle eine Liste der Grauwertbilder
bilder = []
for file in weiss_files:
    img = cv2.imread(file)
    if img is not None:
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY).astype("float32")
        bilder.append(gray)
    else:
        print(f"Bild konnte nicht geladen werden: {file}")

# Überprüfe, ob Bilder erfolgreich geladen wurden
if bilder:
    # Mittelwert der Weißbilder berechnen
    weissbild = np.mean(bilder, axis=0)

    # Dunkelbild abziehen
    dunkelbild = np.load("BildOutput/dunkelbild.npy")
    weissbild_korrigiert = weissbild - dunkelbild

    # Weißbild normalisieren (Mittelwert = 1)
    weissbild_normiert = weissbild_korrigiert / np.mean(weissbild_korrigiert)

    # Weißbild speichern im Ordner "Bildaufnahmen"
    np.save("BildOutput/weissbild.npy", weissbild_normiert)

    # Weißbild kontrastmaximiert darstellen
    plt.imshow(weissbild_normiert, cmap="gray")
    plt.title("Kontrastmaximiertes Weißbild")
    plt.show()
else:
    print("Es wurden keine gültigen Weißbilder gefunden.")
```

```python
gray_values = [gray[:,0:108], gray[:,109:246], gray[:,247:390], gray[:,391:532], gray[:,533:640]]

# Berechnung der Mittelwerte und Standardabweichungen der Grauwertstufe
mittelwerte = []
stdabw = []

# Berechnung der Mittelwerte und Standardabweichungen für jede Graustufenregion
for region in gray_values:
    region_flat = region.flatten()
    mittelwerte.append(np.mean(region_flat))
    stdabw.append(np.std(region_flat))

# Erstellen eines DataFrames mit den Ergebnissen
df = pd.DataFrame({
    'Graustufen-Bereich': [f'Graustufe {i+1}' for i in range(5)],
    'Mittelwert': mittelwerte,
    'Standardabweichung': stdabw
})

# Tabelle anzeigen
display(df)


```

```python
# Grauwertkeil-Bild einlesen und korrigieren

# Versuche, das Grauwertkeil-Bild im Ordner "Bildaufnahmen" zu laden
bild = cv2.imread("Bildaufnahmen/Grauwertkeil_farbe.png")

# Überprüfen, ob das Bild erfolgreich geladen wurde
if bild is None:
    print("Das Bild konnte nicht geladen werden. Überprüfe den Pfad und die Datei.")
else:
    # Das Bild in Graustufen umwandeln
    gray = cv2.cvtColor(bild, cv2.COLOR_BGR2GRAY).astype("float32")

    # Dunkelbild und Weißbild laden
    dunkelbild = np.load("BildOutput/Dunkelbild.npy")
    weissbild = np.load("BildOutput/weissbild.npy")

    # Korrektur des Grauwertkeils durchführen
    korrigiert = (gray - dunkelbild) / weissbild
    korrigiert = np.clip(korrigiert, 0, 255)  # Werte auf den Bereich 0 bis 255 begrenzen

    # Korrigiertes Grauwertkeil-Bild im Ordner "Bildaufnahmen" speichern
    cv2.imwrite("BildOutput/grauwertkeil_korrigiert.png", korrigiert.astype("uint8"))

    # Korrigiertes Bild anzeigen
    plt.imshow(korrigiert, cmap="gray")
    plt.title("Korrigiertes Grauwertkeil-Bild")
    plt.show()

```

```python
#GrauStufen
grayk_values = [gray[:,0:108], gray[:,109:246], gray[:,247:390], gray[:,391:532], gray[:,533:640]]

# Einzelanzeige mit Subplots
plt.figure(figsize=(15, 5))  # Breite und Höhe des gesamten Plots

for i, gray in enumerate(grayk_values):
    plt.subplot(1, 5, i + 1)  # 1 Zeile, 5 Spalten
    plt.imshow(gray, cmap='gray', vmin=0, vmax=255)  # Graustufenanzeige
    plt.title(f'Graustufe {i+1}')
    plt.axis('off')  # Achsen ausblenden

plt.tight_layout()
plt.show()
```

```python
# Berechnung der Mittelwerte und Standardabweichungen der Grauwertstufen

# Überprüfen, ob das korrigierte Bild (korrigiert) existiert
if 'korrigiert' not in locals():
    print("Das Bild wurde noch nicht korrigiert. Bitte führe zuerst die Zelle zum Korrigieren des Grauwertkeils aus.")
else:
    # Die 5 vertikalen Graustufen-Streifen (wie im ersten Code)
    grey_regions = [
        korrigiert[:, 0:108],
        korrigiert[:, 109:246],
        korrigiert[:, 247:390],
        korrigiert[:, 391:532],
        korrigiert[:, 533:640]
    ]

    mittelwerte = []
    stdabw = []

    # Berechnung der Mittelwerte und Standardabweichungen für jede Graustufenregion
    for region in grey_regions:
        region_flat = region.flatten()
        mittelwerte.append(np.mean(region_flat))
        stdabw.append(np.std(region_flat))

    # Erstellen eines DataFrames mit den Ergebnissen
    df = pd.DataFrame({
        'Graustufen-Bereich': [f'Graustufe {i+1}' for i in range(5)],
        'Mittelwert': mittelwerte,
        'Standardabweichung': stdabw
    })

    # Tabelle anzeigen
    display(df)

```

```python
# Sicherstellen, dass die nötigen Bilder vorhanden sind
if 'dunkelbild' not in locals() or 'weissbild' not in locals():
    print("Bitte lade zuerst das Dunkelbild und das Weißbild.")
else:
    # Dynamische Schwellenwerte je nach Wertebereich
    if weissbild.max() <= 1.0:
        stuck_pixel_threshold = 0.99999999
        dead_pixel_threshold = 0.00000001
    else:
        stuck_pixel_threshold = 254
        dead_pixel_threshold = 1

    # Fehlerpixelmasken berechnen
    fehler_dunkel = (dunkelbild > stuck_pixel_threshold)
    fehler_weiss = (weissbild < dead_pixel_threshold)

    n_hot_stuck = np.sum(fehler_dunkel)
    n_dead = np.sum(fehler_weiss)

    print(f"Gefundene stuck/hot pixels: {n_hot_stuck}")
    print(f"Gefundene dead pixels: {n_dead}")
    print(f"Gesamtanzahl fehlerhafter Pixel: {n_hot_stuck + n_dead}")

    # Plot vorbereiten (immer 2 Bilder)
    plt.figure(figsize=(12, 5))

    # 1. Plot: Dunkelbild (immer anzeigen, Fehler optional markieren)
    plt.subplot(1, 2, 1)
    plt.imshow(dunkelbild, cmap='gray', vmin=0, vmax=255)
    if n_hot_stuck > 0:
        y, x = np.where(fehler_dunkel)
        plt.scatter(x, y, color='red', s=1, label='hot/stuck pixel')
        plt.legend()
    plt.title("Dunkelbild mit hot/stuck pixels")
    plt.axis('off')

    # 2. Plot: Weißbild mit dead pixels
    plt.subplot(1, 2, 2)
    plt.imshow(weissbild, cmap='gray', vmin=0, vmax=255)
    if n_dead > 0:
        y, x = np.where(fehler_weiss)
        plt.scatter(x, y, color='blue', s=1, label='dead pixel')
        plt.legend()
    plt.title("Weißbild mit dead pixels")
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```
